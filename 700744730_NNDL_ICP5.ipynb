```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score
```

```python
from google.colab import files
# Uploading the glass data CSV file
uploaded = files.upload()
import os
# we can see if the file uploaded in the current directory
current_directory = os.getcwd()
directory_contents = os.listdir(current_directory)
print(directory_contents)
```

```python
# Load the glass dataset
glass_data = pd.read_csv("glass.csv")
glass_data.head()
```

```python
# Split the dataset into features and target
x_train = glass_data.drop("Type", axis=1)
y_train = glass_data['Type']
x_train.head()
```

```python
# Split the dataset into training and testing parts
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.2, random_state=0)
```

```python
# Fitting the Naive Bayes model on the training data
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
# Predict the target on the test data
y_pred = gnb.predict(x_test)
```

```python
# Evaluate the model
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Classification Report:\n",classification_report(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Split tha dataset into features and target
x_train = glass_data.drop("Type", axis=1)
y_train = glass_data['Type']
```

```python
# Split the dataset into training and testing parts
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.2, random_state=0)
```

```python
# Fitting the linear SVM model on the training data
svc = SVC()
svc.fit(x_train, y_train)

```

```python
# Predict the target on the test data
y_pred = svc.predict(x_test)
```

```python
# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
